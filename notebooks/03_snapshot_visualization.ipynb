# 03 — Snapshot Visualization

Grid of CFD and rCFD frames at matched time steps.

```python
import pathlib, matplotlib.pyplot as plt
from PIL import Image
DATA = pathlib.Path('..') / 'data'
indices = [60, 180, 300, 420, 540]
fig, axes = plt.subplots(2, len(indices), figsize=(3*len(indices), 6))
for j, i in enumerate(indices):
    axes[0, j].imshow(Image.open(DATA / 'cfd_snapshots'  / f'CFD_T_path_{i:04d}.jpg'))
    axes[0, j].set_title(f'CFD t={i*0.2:.1f}s'); axes[0, j].axis('off')
    axes[1, j].imshow(Image.open(DATA / 'rcfd_snapshots' / f'rCFD_BJet_{i:04d}.jpg'))
    axes[1, j].set_title(f'rCFD t={i*0.2:.1f}s'); axes[1, j].axis('off')
plt.tight_layout(); plt.show()
```
