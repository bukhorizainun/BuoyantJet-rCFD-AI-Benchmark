# 01 — Dataset Overview

Sanity-check the local CFD/rCFD dataset and visualize a few snapshots.

Expected layout: see `data/README.md`.

```python
import sys, pathlib
sys.path.append(str(pathlib.Path('..').resolve() / 'scripts'))
from cfdio import list_snapshots

DATA = pathlib.Path('..') / 'data'
cfd = list_snapshots(DATA / 'cfd_snapshots', 'CFD_T_path_*.jpg')
rcfd = list_snapshots(DATA / 'rcfd_snapshots', 'rCFD_BJet_*.jpg')
print('CFD frames :', len(cfd))
print('rCFD frames:', len(rcfd))
```

```python
import matplotlib.pyplot as plt
from PIL import Image
fig, axes = plt.subplots(2, 3, figsize=(10, 6))
for ax, p in zip(axes[0], cfd[:3]):
    ax.imshow(Image.open(p)); ax.set_title(p.name); ax.axis('off')
for ax, p in zip(axes[1], rcfd[:3]):
    ax.imshow(Image.open(p)); ax.set_title(p.name); ax.axis('off')
plt.tight_layout(); plt.show()
```
